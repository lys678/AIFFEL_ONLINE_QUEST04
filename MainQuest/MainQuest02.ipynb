```python
import random

class Aiffel:
    core_list = {
        '김  건': 0, '김경민': 0, '김소연': 0, '김영신': 0, '박근수': 0, '박영수': 0,
        '박영준': 0, '서원영': 0, '신지만': 0, '심지안': 0, '이규상': 0, '이윤상': 0,
        '이진영': 0, '임현석': 0, '전요한': 0, '전재영': 0, '조대희': 0
    }
    core_order = 1

    def __init__(self, name):
        self.name = name  # 이름
        self.admission_year = 2023  # 입학연도 고정
        self.cohort = 25  # 기수 고정
        self.course = "코어"  # 과정명 고정
        self.student_id = self.generate_student_id()  # 학번 생성
        self.quest_score = 0  # 퀘스트 점수

    def generate_student_id(self):
        # 연도 마지막 두 자리 추출, 100으로 나눈 후 나머지를 이용함
        year = self.admission_year % 100

        # 기수 추출
        cohort_number = self.cohort

        # 과정 종류 추출
        if self.course=="코어":
            course_type=2
        else:
            course_type=1

        # 학번 생성
        order_number = self.core_order
        student_id = f"{year:02d}{cohort_number:02d}{course_type}{order_number:03d}"

        # 가나다순 번호 증가
        Aiffel.core_order += 1

        return student_id

    def score(self, points):
        if 1 <= points <= 3:
            self.quest_score += points
        else:
            print("퀘스트 점수는 1부터 3까지만 입력 가능합니다.")

    def penalty(self, points):
        if 1 <= points <= 3:
            self.quest_score -= points
        else:
            print("퀘스트 점수는 1부터 3까지만 입력 가능합니다.")

    @classmethod
    def get_id_num(cls):
        return len(cls.core_list)

    def give_score_penalty(cls, student_name, score_points, penalty_points):
        for student in cls.core_list.keys():
            if student == student_name:
                cls.core_list[student] += score_points
                cls.core_list[student] -= penalty_points

    @classmethod
    def remove_student(cls, student_name):
        if student_name in cls.core_list:
            del cls.core_list[student_name]

    @classmethod
    def reset_core_order(cls):
        cls.core_order = 1


# 학생 인스턴스 생성
students = [Aiffel(name) for name in Aiffel.core_list.keys()]

# 특정 학생을 고르고 점수와 패널티를 부여
student = Aiffel("김건")
student.score(2)  # 점수 2를 주는 경우
student.penalty(1)  # 점수 1을 빼는 경우

print(student.quest_score)  # 출력: 1

# 모든 학생 정보 출력
print("=== 모든 학생 정보 ===")
for student in students:
    print(f"이름: {student.name}, 학번: {student.student_id[:5]}{student.student_id[6:]}, 퀘스트 점수: {student.quest_score}")

# 생성된 학번의 개수 출력
print(f"생성된 학번의 개수: {Aiffel.get_id_num()}")

# 학생 삭제
Aiffel.remove_student("김경민")
Aiffel.remove_student("김영신")

# core_order 초기화
Aiffel.reset_core_order()

# 학생 정보가 갱신된 students 리스트 재생성
students = [Aiffel(name) for name in Aiffel.core_list.keys()]

# 모든 학생 정보 출력 (갱신 전 명단)
print("=== 갱신 전 학생 정보 ===")
for student in students:
    print(f"이름: {student.name}, 학번: {student.student_id[:5]}{student.student_id[6:]}, 퀘스트 점수: {student.quest_score}")

# 생성된 학번의 개수 출력 (갱신된 명단)
print(f"생성된 학번의 개수: {Aiffel.get_id_num()}")



class Aiffel_Group(Aiffel):
    def __init__(self, name):
        super().__init__(name)
        self.group = None  # 소속 그룹

    @staticmethod
    def grouping(students):
        groups = ['상생', '신뢰', '열정', '이끔', '성장', '가치', '공유', '확산']
        random.shuffle(groups)

        for i, student in enumerate(students):
            student.group = groups[i % len(groups)]

    def print_grouping(self):
        print(f"이름: {self.name}, 그룹: {self.group}")

# 학생 인스턴스 생성
students = [Aiffel_Group(name) for name in Aiffel.core_list.keys()]

# 학생 그룹핑
Aiffel_Group.grouping(students)

# 그룹핑 결과 출력
print("=== 그룹핑 결과 ===")
for student in students:
    student.print_grouping()

class Aiffel_Guild(Aiffel):
    guilds = ['힌튼', '알트만', '응', '하사비스']

    def __init__(self, name):
        super().__init__(name)
        self.guild = None  # 소속 길드

    @staticmethod
    def grouping(students):
        random.shuffle(Aiffel_Guild.guilds)

        for i, student in enumerate(students):
            student.guild = Aiffel_Guild.guilds[i % len(Aiffel_Guild.guilds)]

    def print_guilding(self):
        print(f"이름: {self.name}, 길드: {self.guild}")

# 학생 인스턴스 생성
students = [Aiffel_Guild(name) for name in Aiffel.core_list.keys()]


class Aiffel_Guild(Aiffel):
    guilds = ['힌튼', '알트만', '응', '하사비스']
    guild_scores = {guild: 0 for guild in guilds}

    def __init__(self, name):
        super().__init__(name)
        self.guild = None  # 소속 길드

    @staticmethod
    def grouping(students):
        random.shuffle(Aiffel_Guild.guilds)

        for i, student in enumerate(students):
            student.guild = Aiffel_Guild.guilds[i % len(Aiffel_Guild.guilds)]

    def guild_score(self, guild_name, score):
        if guild_name in Aiffel_Guild.guild_scores:
            Aiffel_Guild.guild_scores[guild_name] += score
            self.score(score)  # 개별 그루의 점수도 증감

    def print_guilding(self):
        print(f"이름: {self.name}, 길드: {self.guild}")

# 학생 인스턴스 생성
students = [Aiffel_Guild(name) for name in Aiffel.core_list.keys()]

# 학생 길드 그룹핑
Aiffel_Guild.grouping(students)

# 그룹핑 결과 출력
print("=== 길드 그룹핑 결과 ===")
for student in students:
    student.print_guilding()

# 길드 점수 입력 및 출력
students[0].guild_score("힌튼", 2)
students[1].guild_score("알트만", 3)
students[2].guild_score("응", 1)

print("\n=== 길드 점수 ===")
for guild, score in Aiffel_Guild.guild_scores.items():
    print(f"{guild}: {score}")

# 길드 점수 차감 및 개별 그루 점수 출력
students[0].guild_score("힌튼", -3)
students[1].guild_score("알트만", -1)
students[3].guild_score("하사비스", 2)

print("\n=== 길드 점수 ===")
for guild, score in Aiffel_Guild.guild_scores.items():
    print(f"{guild}: {score}")

# 개별 그루 점수 출력
print("\n=== 개별 그루 점수 ===")
for student in students:
    print(f"이름: {student.name}, 그루: {student.student_id[:5]}{student.student_id[6:]}, 점수: {student.quest_score}")

```
